```python
def clean_data(x):
    if type(x) == str:
        return float(x[:-1])/100
    elif x is None:
        return 0
    else:
        return x
        
```

```python
#find avg velocity by multiplying perc. pitch used by it's avg velocity and adding them all up
def get_avg_speed(player):
    
    fast = player['FB%'] *player['FBv']
    slider = player['SL%'] *player['SLv']
    cutter = player['CT%'] *player['CTv']
    curve = player['CB%'] *player['CBv']
    change = player['CH%'] *player['CHv']
    splitter = player['SF%'] *player['SFv']
    knuckle = player['KN%'] *player['KNv']
    
    avg_speed = fast+ slider+cutter+curve+change+splitter+knuckle 
#     print(avg_speed)
    return round(avg_speed,2)


# find out how many pitches in a pitchers repertoire
def dif_pitches(player): 
    pitch_types = 0
    
    if player['FBv'] !=0.0:
        pitch_types += 1
    if player['SLv'] !=0.0:
        pitch_types += 1 
    if player['CTv'] !=0.0:
        pitch_types += 1   
    if player['CBv'] !=0.0:
        pitch_types += 1 
    if player['CHv'] !=0.0:
        pitch_types += 1
    if player['SFv'] !=0.0:
        pitch_types += 1
    if player['KNv'] !=0.0:
        pitch_types += 1
    return pitch_types
```

```python
pitches_2012['FB%'] = pitches_2012['FB%'].apply(clean_data)
pitches_2012['FBv'] = pitches_2012['FBv'].apply(clean_data)
pitches_2012['SL%'] = pitches_2012['SL%'].apply(clean_data)
pitches_2012['SLv'] = pitches_2012['SLv'].apply(clean_data)
pitches_2012['CT%'] = pitches_2012['CT%'].apply(clean_data)
pitches_2012['CTv'] = pitches_2012['CTv'].apply(clean_data)
pitches_2012['CB%'] = pitches_2012['CB%'].apply(clean_data)
pitches_2012['CBv'] = pitches_2012['CBv'].apply(clean_data)
pitches_2012['CH%'] = pitches_2012['CH%'].apply(clean_data)
pitches_2012['CHv'] = pitches_2012['CHv'].apply(clean_data)
pitches_2012['SF%'] = pitches_2012['SF%'].apply(clean_data)
pitches_2012['SFv'] = pitches_2012['SFv'].apply(clean_data)
pitches_2012['KN%'] = pitches_2012['KN%'].apply(clean_data)
pitches_2012['KNv'] = pitches_2012['KNv'].apply(clean_data)
pitches_2012['XX%'] = pitches_2012['XX%'].apply(clean_data)
```
